```python
import pandas as pd

div_user_norm = pd.read_csv('dived_user_gp_norm.csv', index_col=[0], header=[0, 1, 2, 3], skipinitialspace=True, tupleize_cols=True)
div_file_norm = pd.read_csv('dived_file_gp_norm.csv', index_col=[0], header=[0, 1, 2, 3], skipinitialspace=True, tupleize_cols=True)
div_user_norm
```

```python
def save_barh(df, name, figsize=(10, 120)):
    figure = df.plot.barh(stacked=True, figsize=figsize).get_figure()
    figure.subplots_adjust(left=0.2)
    figure.savefig(name)
    plt.clf()
```

```python
for index, row in acc_group.iterrows():
    acc_df = row.unstack().fillna(0)
    indexer = acc_df.sum(1).argsort()
    sorted_acc_df = acc_df.take(indexer)
    save_barh(sorted_acc_df.tail(20), 'plot/each_acc_dividedFile/sorted_acc_df_%f_top20.png' % index, figsize=(10, 10))
    save_barh(acc_df.tail(20), 'plot/each_acc_dividedFile/acc_df_%f_top20.png' % index, figsize=(10, 10))
```

```python
def plot_df(basename, label, srs, top=20, ext='png'):
    df = srs.unstack().fillna(0)
    indexer = df.sum(1).argsort()
    sorted_df = df.take(indexer)
    save_barh(sorted_df.tail(20), basename + '_top20_sorted_%s.%s' % (label, ext), figsize=(10, 10))
    save_barh(sorted_df, basename + '_sorted_%s.%s' % (label, ext), figsize=(10, 120))
    save_barh(df, basename + '_%s.%s' % (label, ext), figsize=(10, 120))
```

```python
for rating, srs in div_file_norm.sum().groupby(level=3, group_keys=False):
    srs.index = srs.index.droplevel(3)
    label = rating.replace(' ', '')
    plot_df('plot/each_rating_divFile/rating', label, srs)
```
