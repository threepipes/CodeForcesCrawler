```python
import os
import pymysql
import pandas.io.sql as sql
import pandas as pd

connection = pymysql.connect(host='localhost',
                            user=os.getenv('CF_DB_USER', 'root'),
                            password=os.getenv('MYSQL_PASS', 'pass'),
                            db=os.getenv('CF_DB', 'testcf'),
                            charset='utf8mb4',
                            cursorclass=pymysql.cursors.DictCursor)
connection.commit()
```

```python
df_prob = sql.read_sql("select * from modification where problem_id='213B'", connection)
df_user = sql.read_sql("select * from participant", connection)
df_acc = sql.read_sql("select * from acceptance", connection)
```

```python
df_prob
```

```python
merged = df_prob.merge(df_user).merge(df_acc)
```

```python
merged
```

```python
rating_split = [
    1100, 1200, 1300, 1400, 1500, 1600, 1700, 1900, 2200, 9999
]
rating_split_bin = [-1000] + rating_split
rating_label = ['%d - %d' % (s + 1, t) for s, t in zip(rating_split_bin, rating_split_bin[1:])]
rating_label[0] = ' - 1100'
rating_label[-1] = '2200 - '
```

```python
rating_bin = pd.cut(merged.rating, rating_split_bin, labels=rating_label)
rating_bin
```

```python
merged['rating_bin'] = rating_bin
merged
```

```python
acceptance_split = [0.2 * i for i in range(6)]
merged['acceptance_bin'] = pd.cut(merged.acceptance_rate, acceptance_split)
merged
```

```python
grouped = merged.groupby(['block_type', 'node_type', 'modification_type', 'rating_bin']).size()
grouped
```

```python
prob_info = merged.groupby(['rating_bin']).file_name.agg(lambda x: x.nunique())
prob_info
```

```python
prob_info = merged.groupby(['rating_bin']).user_name.agg(lambda x: x.nunique())
prob_info
```
